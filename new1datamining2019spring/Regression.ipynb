```python
import pandas as pd
import numpy as np
```

```python
# 读入数据
data_train = pd.read_csv('./trainSet.csv')
data_train
```

```python
# 查看相关性
import seaborn as sns
import matplotlib.pyplot as plt
corrmat = data_train.corr()
print(corrmat.iloc[-1, :])
plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=0.8, square=True)
plt.show()
```

```python
# 提取特征
X = data_train.iloc[:, 0:-1]
X = X.drop(columns=['F7' , 'F22', 'F26'])
print(X)
y = data_train.iloc[:, -1]
print(X.shape)
print(y.shape)
```

```python
# 建立LR模型
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='liblinear', C=1, random_state=0)
print(X.shape)
clf.fit(X, y.astype('int'))
clf
```

```python
# 查看模型参数
print(X.columns.shape)
print(clf.coef_.T)
pd.DataFrame({'Attr': list(X.columns)[0:-1], "Coef": list(clf.coef_.T)})
```

```python
# 预测结果
test_data = pd.read_csv('./test set.csv')
#test_data
print(test_data.shape)
predictions = clf.predict(test_data)
n = len(predictions)
result = pd.DataFrame({'ID': list(range(1, n+1)), 'Predicted': predictions.astype(np.int32)})
result.to_csv('./result.csv', index=False)
```

```python
# CV测试
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
seed = 7
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)

# Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

print(X_train_std.shape)
model = LogisticRegression(solver='sag', random_state=0)
model.fit(X_train_std, y_train)
```

```python
y_pred = model.predict(X_test_std)
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test, predictions)
print('Accuracy: %.2f%%' % (accuracy * 100.0))
```
