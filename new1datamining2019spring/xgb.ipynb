```python
import numpy as np
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data_train = pd.read_csv('./trainSet.csv')
data_train
```

```python
X = data_train.iloc[:, 0:-1]
X = X.drop(columns=['F7','F22','F26'])
y = data_train.iloc[:, -1]
seed = 7
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
# Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
model = XGBClassifier()
model.fit(X_train_std, y_train)
```

```python
y_pred = model.predict(X_test_std)
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test, predictions)
print('Accuracy: %.2f%%' % (accuracy * 100.0))
```

```python
X = data_train.iloc[:, 0:-1]
X = X.drop(columns=['F7','F22','F26'])
y = data_train.iloc[:, -1]

# 预测结果
test_data = pd.read_csv('./test set.csv')
#test_data
test_data = test_data.drop(columns=['F7','F22','F26'])
print(test_data.shape)

# Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
X_train_std = sc.transform(X)
X_test_std = sc.transform(test_data)

xgbmodel = XGBClassifier()
xgbmodel.fit(X_train_std, y)
```

```python
predictions = xgbmodel.predict(X_test_std)
n = len(predictions)
result = pd.DataFrame({'ID': list(range(1, n+1)), 'Predicted': predictions.astype(np.int32)})
result.to_csv('./result_xgb.csv', index=False)
```

```python
from xgboost import plot_importance
from matplotlib import pyplot
plot_importance(model)
pyplot.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
```

```python
model = XGBClassifier()
learning_rate = [0.0001, 0.001, 0.01, 0.1]
param_grid = dict(learning_rate=learning_rate)
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
grid_search = GridSearchCV(model, param_grid, scoring="neg_log_loss", n_jobs=-1, cv=kfold)
grid_result = grid_search.fit(X, y)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
